## Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

## Load the Data

```python
df1 = pd.read_csv("/content/prevalence-by-mental-and-substance-use-disorder.csv")
df2 = pd.read_csv("/content/mental-and-substance-use-as-share-of-disease.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
data = pd.merge(df1,df2)
data.head(11)
```

## Data Cleaning

```python
data.isnull().sum()
```

```python
data.drop('Code',axis=1,inplace=True)
```

```python
data.head(11)
```

```python
data.size,data.shape
```

```python
data.set_axis(['Country','Year','Schizophrenia','Bipolar_disorder','Eating_disorder','Anxiety','drug_usage','depression','alcohal','mental_fitness'],axis='columns',inplace=True)
data.head(11)
```

## Data Visualization

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.corr(),annot=True,cmap='Reds')
plt.plot()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in data.columns:
  if data[i].dtype == 'object':
    data[i]=l.fit_transform(data[i])
```

```python
data.shape
```

## Split Data

```python
X = data.drop('mental_fitness',axis=1)
y = data['mental_fitness']

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y, test_size=.20, random_state=2)
```

```python
print("xtrain: ",xtrain.shape)
print("xtest: ",xtest.shape)
print("ytrain: ",ytrain.shape)
print("ytest: ",ytest.shape)
```

## Model Training

fitting data

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
rf = RandomForestRegressor()
rf.fit(xtrain,ytrain)
```

model evaluation for training set

```python
ytrain_pred = rf.predict(xtrain)
```

mean squared error(Observed value)

```python
mse = mean_squared_error(ytrain,ytrain_pred)
```

Root mean square error

```python
rmse = (np.sqrt(mean_squared_error(ytrain,ytrain_pred)))
```

R2 (goodness of fit of a model)

```python
r2 = r2_score(ytrain,ytrain_pred)
```

```python
print("The RandomForestRegressor model performance for training set")
print("..........................")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))
```
